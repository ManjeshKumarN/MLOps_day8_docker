```python
# login to docker hub, install docker desktop then only in local system docker commands runs
# docker hub acts as remote repository for Docker images, from which anyone can download the image and use it

# docker image -- on running --> becomes docker container
# docker image is built using Dockerfile

# in github codespaces Docker is pre installed
# docker build -t flask-docker-demo . -> to build as docker image from the Docker file, name of the docker file is flask-docker-demo
# docker run hello-world -> 'hello-world' is image name , to pull( if images is not present in local ) and run the image 
# docker pull <image_name>:<tag_name> -> docker pull hello-world:nanoserver-ltsc2025 , to pull the image with respective image
# docker ps  -> to list the running containers
# docker ps -a  -> to list the exited containers, stopped containers , which can be started again by "docker start" command
```

```python
# port mapping
# docker run -p 8000:5000 flask-docker-demo --> docker run -p host_port:container_port flask-docker-demo
# docker run -d -p 6000:5000 flask-docker-demo -> to run the docker image in detach mode , i.e continer will run in the background, with closing the terminal

# docker stop <container>  -> to the stop the running conatiner, can use container_id, container name
# docker start <container> -> to start the stopped conatiner ( conatiners under docker rm , can be started)
```

```python
# docker tagging , tagging an image for pushing to docker hub repository

# docker tag <image> <repo>:<tag>  - > Tag an image for pushing to a registry

# docker tag flask-docker-demo manjeshkumar/flask-docker-demo:newest  # this creates an image with another name

# it is not a copy of the original image it is just renaming, the image id of both tagged and original images are same

# docker push <repo>:<tag> -> Push an image to a remote registry
# docker push manjeshkumar/flask-docker-demo:newest # push the renamed image 

# docker login -u manjeshkumar -> before pushing need to login
# enter the personal access token 

# complete flow  build docker images  tag images  push the image to repository  pull from repository
```
